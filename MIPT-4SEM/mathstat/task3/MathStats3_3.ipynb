```python
# Импортируем всякие библиотеки
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt

%matplotlib inline

GRAPH_ALPHA = 0.5
GRAPH_LINEWIDTH = 0.1

SIZE = 1000
NUMBER_OF_SAMPLES = 500

thetaDelta = 0.01

disEstimate = []

currTheta = 0
while currTheta < 1:
    disEstimate.append((currTheta * (1 - currTheta)) / (SIZE))
    currTheta += thetaDelta

grid = np.linspace(0, 1, 100)

plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
#plt.axis([0, SIZE, 0, 0.00002])
plt.plot(grid, disEstimate, color='green', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation')
plt.legend()  # Ставим легенду
plt.title('bernulli')
plt.show()  # Показываем график

disEstimate = []
currTheta = 0
while currTheta < 1:
    estimates = []
    for i in range(NUMBER_OF_SAMPLES):
        sample = sst.binom.rvs(1, currTheta, SIZE)
        estimates.append(np.mean(sample))
    disEstimate.append(np.var(estimates))
    currTheta += thetaDelta
    
plt.figure(figsize=(18, 5))  # Инициализируем график и задаем его размеры
#plt.axis([0, SIZE, 0, 0.00002])
plt.plot(grid, disEstimate, color='red', alpha=1, 
         label=r'$D_{\theta}(\hat{\theta})$ - estimation')
plt.legend()  # Ставим легенду
plt.title('bernulli')
plt.show()  # Показываем график
```
