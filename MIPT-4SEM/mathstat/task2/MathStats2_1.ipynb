Рассмотрим нормальное распределение

```python
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt

%matplotlib inline

GRAPH_ALPHA = 0.5
GRAPH_LINEWIDTH = 0.1

SIZE = 1000
SCALE = 3
mu = 2
n = 3
p = 0.5

sampleNorm = sst.norm.rvs(size=SIZE, loc=1, scale=SCALE)

sampleNormSqr = []

for i in range(SIZE):
    sampleNormSqr.append(sampleNorm[i] ** 2)

grid = np.linspace(-10, 10, SIZE)
scaleA = [0]
scaleSigma = [0]

for i in range(1, SIZE):
    scaleA.append(np.mean(sampleNorm[:i]))
    scaleSigma.append(np.mean(sampleNormSqr[:i]) - (scaleA[i]) ** 2)
    
plt.figure(figsize=(18, 5))
plt.plot(grid, scaleA, color='red', alpha = 1)
plt.plot(grid, np.ones(SIZE), color='blue', alpha = 1)
plt.title(r'Normal: a with moments method') 
plt.show()

plt.figure(figsize=(18, 5))
plt.plot(grid, scaleSigma, color='green', alpha = 1)
plt.plot(grid, np.ones(SIZE) * (SCALE ** 2), color='blue', alpha = 1)
plt.title(r'Normal: $\sigma^2$ with moments method') 
plt.show()
```

```python
import numpy as np
import scipy.stats as sst
import matplotlib.pyplot as plt

%matplotlib inline

GRAPH_ALPHA = 0.5
GRAPH_LINEWIDTH = 0.1

SIZE = 1000
SCALE = 3
mu = 2
n = 3
p = 0.5

sampleNorm = sst.norm.rvs(size=SIZE, loc=1, scale=SCALE)
sampleR = sst.uniform.rvs(size=SIZE, scale=SCALE)
samplePois = sst.poisson.rvs(mu, size=SIZE)
sampleBin = sst.binom.rvs(n, p, size=SIZE)
sampleGeom = sst.geom.rvs(p, size=SIZE)

sampleNormSqr = []
sampleRSqr = []
samplePoisSqr = []
sampleBinSqr = []
sampleGeomSqr = []

for i in range(SIZE):
    sampleNormSqr.append(sampleNorm[i] ** 2)
    sampleRSqr.append(sampleR[i] ** 2)
    samplePoisSqr.append(samplePois[i] ** 2)
    sampleBinSqr.append(sampleBin[i] ** 2)
    sampleGeomSqr.append(sampleGeom[i] ** 2)

grid = np.linspace(-10, 10, SIZE)
scaleA = [0]
scaleSigma = [0]
for i in range(1, SIZE):
    scaleA.append(np.mean(sampleNorm[:i]))
    scaleSigma.append(np.mean(sampleNormSqr[:i]) - (scaleA[i]) ** 2)
    
plt.figure(figsize=(18, 5))
plt.plot(grid, scaleA, color='red', alpha = 1)
plt.plot(grid, np.ones(SIZE), color='blue', alpha = 1)
plt.title(r'Normal: a = ') 
plt.show()

plt.figure(figsize=(18, 5))
plt.plot(grid, scaleSigma, color='green', alpha = 1)
plt.plot(grid, np.ones(SIZE) * (SCALE ** 2), color='blue', alpha = 1)
plt.title(r'Normal: $\sigma^2$ = ') 
plt.show()
```
