```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

```python
N = 10000
theta = 1.0

# k - константа из оценки
# scale - размер
def showGraphs(k, scale):
    global theta
    sample = sts.expon(theta).rvs(N)
    
    x = np.arange(1, N + 1)
    y = np.zeros(N)

    for n in range(1, N):
        samplePart = sample[:n]
        y[n] = abs(theta - (float(math.factorial(k)) \
                            / (samplePart ** k).mean()) ** (1.0 / k))
   
    plt.figure(figsize=(9, 5))
    plt.plot(x, y, label='${(k!/\overline{X^{k}})}^{1/k}$')
    plt.legend()
    plt.ylim((0, scale))
    plt.xlabel('n')
    plt.ylabel('$|\\theta - estimation|$')
    plt.title('k=' + str(k))
    plt.show()
    print "Delta:", y[-1]
```

```python
for k in range(1, 6):
    showGraphs(k, 1)
```

```python
for k in range(6, 15, 2):
    showGraphs(k, 1)
```

```python
for k in range(15, 36, 5):
    showGraphs(k, 1)
```
